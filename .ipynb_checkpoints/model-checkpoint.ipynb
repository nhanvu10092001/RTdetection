```python
def create_file(json_dir, save_dir, mode):
    import json
    import shutil
    with open(json_dir) as file:
        annotation  = json.load(file)
        file.close()
        
    for item in annotation['annotations']:
        if(item["category_id"] == 1):

            for image_info in annotation['images']:
                if image_info["id"] == item["image_id"]:
                    file_name = image_info['file_name']
                    top_left_x, top_left_y, box_width, box_height = item["bbox"]
                    width = image_info["width"]
                    height = image_info["height"]
                    shutil.copy2("/home/nhanvu10092001/fiftyone/coco-2017/{}/data/{}".format(mode, file_name), '{}/images/{}'.format(save_dir, mode))
                    with open("{}/labels/{}/{}.txt".format(save_dir, mode, file_name.split(".")[0]), "a") as F:
                        
                        F.writelines("{} {} {} {} {}\n".format(0, (top_left_x + box_width/2)/width, (top_left_y + box_height/2)/height, box_width/width, box_height/height))
                        F.close()
        
                    
```

```python

create_file(TRAIN_ANNOTATIONS_PATH_JSON, save_dir = "/home/nhanvu10092001/Seminar/data",mode = "train")
create_file(VALID_ANNOTATIONS_PATH_JSON, save_dir = "/home/nhanvu10092001/Seminar/data",mode = "validation")
```

```python
import cv2
from ultralytics import YOLO


# Load the YOLOv8 model
#model = YOLO('yolov8n.pt')

# Open the video file

cap = cv2.VideoCapture(500, cv2.CAP_V4L2)

# Loop through the video frames
while cap.isOpened():
    # Read a frame from the video
    success, frame = cap.read()
   

    if success:
        # Run YOLOv8 inference on the frame
        results = model(frame)

        # Visualize the results on the frame
        annotated_frame = results[0].plot()

        # Display the annotated frame
        cv2.imshow("webcam", annotated_frame)

        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
cap.release()
```
